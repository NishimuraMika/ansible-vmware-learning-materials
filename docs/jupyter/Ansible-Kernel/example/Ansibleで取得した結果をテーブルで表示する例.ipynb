```ansible
#python
import pandas as pd
import json

# 空のテーブルセルを用意
vm_name = []
vm_ip = []
vm_power = []
vm_uuid = []
on_esxi = []

# セルにデータを入れていく
vms = vm_list['virtual_machines']
for vm in sorted(vms.keys()):
    vm_name.append(vm)
    vm_ip.append(vms[vm]['ip_address'])
    vm_power.append(vms[vm]['power_state'])
    vm_uuid.append(vms[vm]['uuid'])
    on_esxi.append(vms[vm]['esxi_hostname'])

# tableで表示
table = pd.DataFrame({'VM Name': vm_name,
                                        'Network': vm_ip,
                                        'PowerState': vm_power,
                                        'uuid': vm_uuid,
                                        'ESXi': on_esxi,
})

table
```
